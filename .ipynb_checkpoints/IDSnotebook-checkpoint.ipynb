```python
features = pd.read_csv('UNSW_NB15_features.csv')
```

```python
features.head()
```

```python
features['Type '] = features['Type '].str.lower()
```

```python
# selecting column names of all data types
nominal_names = features['Name'][features['Type ']=='nominal']
integer_names = features['Name'][features['Type ']=='integer']
binary_names = features['Name'][features['Type ']=='binary']
float_names = features['Name'][features['Type ']=='float']
```

```python
# selecting common column names from dataset and feature dataset
cols = data.columns
nominal_names = cols.intersection(nominal_names)
integer_names = cols.intersection(integer_names)
binary_names = cols.intersection(binary_names)
float_names = cols.intersection(float_names)
```

```python
# Converting integer columns to numeric
for c in integer_names:
  pd.to_numeric(data[c])
```

```python
# Converting binary columns to numeric
for c in binary_names:
  pd.to_numeric(data[c])
```

```python
# Converting float columns to numeric
for c in float_names:
  pd.to_numeric(data[c])
```

```python
num_col = data.select_dtypes(include='number').columns

# selecting categorical data attributes
cat_col = data.columns.difference(num_col)
cat_col = cat_col[1:]
cat_col
```

```python
# creating a dataframe with only categorical attributes
data_cat = data[cat_col].copy()
data_cat.head()
```

```python
# one-hot-encoding categorical attributes using pandas.get_dummies() function
data_cat = pd.get_dummies(data_cat,columns=cat_col)
```

```python
data_cat.head()
```

```python
data = pd.concat([data, data_cat],axis=1)
```

```python
data.drop(columns=cat_col,inplace=True)
```

```python
data.shape
```

```python
y = data["label"]
```

```python
# Scale numerical features
scaler = StandardScaler()
data_cat = scaler.fit_transform(data_cat)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data_cat, y, test_size=0.2, random_state=42
)
```

```python
# Feature selection using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
feature_importances = rf.feature_importances_
```

```python
# Select top 20 features 
top_feature_indices = rf.feature_importances_.argsort()[-20:]
X_train = X_train[:, top_feature_indices]  # Select top features from X_train using indices
X_test = X_test[:, top_feature_indices]   # Select top features from X_test using indices
```

```python
# Train the Multi-class SVM model
svm = SVC(kernel="rbf", gamma="auto", probability=True)  # Consider adjusting hyperparameters
svm.fit(X_train, y_train)
```
